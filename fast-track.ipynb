### 1. Расширенный EDA

```python
import pandas as pd
import numpy as np
from scipy import stats as st
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance


sns.set_theme(style="whitegrid")


median_price = data_train['price'].median()
data_train['price_group'] = np.where(data_train['price'] < median_price, 'low', 'high')

alpha = 0.05
results = {}

for i, feature in enumerate(cfg.features):

    results[feature] = {}

    print(f'{i+1}. Feature: {feature}')

    print()

    sample = data_train.loc[:, feature]
    sample_low_price = data_train.loc[data_train['price_group'] == 'low', feature]
    sample_high_price = data_train.loc[data_train['price_group'] == 'high', feature]

    if not (sample.dtype == np.float64 or sample.dtype == np.int64):
        continue

    # Mann-Whitney U test (non-parametric)
    # Motivation: Water resistance might not be normally distributed
    statistic, p_value = st.mannwhitneyu(sample_low_price, sample_high_price)
    reject = bool(p_value < alpha)
    results[feature]['different'] = reject
    print('Mann-Whitney U Test')
    print('H_0: The distribution of the underlying samples is the same.')
    print(f'Test Statistic: {statistic}')
    print(f'P-value: {p_value:.2f} vs {alpha:.2f}')
    if reject:
        print('Result: H_0 is rejected, the distributions are to be assumed different.')
    else:
        print('Result: H_0 is not rejected, the distributions are to be assumed the same.')

    print()

    statistic, p_value = st.normaltest(sample)
    reject = bool(p_value < alpha)
    results[feature]['normal'] = not reject
    print("D'Agostino's K-squared Test")
    print('H_0: The underlying sample follows a normal distribution.')
    print('H_0: The distribution of the underlying samples is the same.')
    print(f'Test Statistic: {statistic}')
    print(f'P-value: {p_value:.2f} vs {alpha:.2f}')
    if reject:
        print('Result: H_0 is rejected, the sample is to not be assumed normally distributed.')
    else:
        print('Result: H_0 is not rejected, the sample is to be assumed normally distributed.')

    print()

    fig, axes = plt.subplots(1, 2, figsize=(12, 3))

    axes[0].set_title("KDE Plot")
    sns.kdeplot(data=data_train, x=feature, ax=axes[0])

    axes[1].set_title("Violin Plot")
    sns.violinplot(data=data_train, x='price_group', y=feature, hue='price_group', ax=axes[1])

    plt.show()

    print()
```

### Вид отчетности

```python
import numpy as np
```

```python
def k_fold_cv(X, y, k=5):

    n_samples = len(X)
    ids = np.arange(n_samples)
    np.random.shuffle(ids)

    fold_sizes = (n_samples // k) * np.ones(k, dtype=int)
    fold_sizes[:n_samples % k] += 1

    current = 0
    folds = []
    for fold_size in fold_sizes:
        start, stop = current, current + fold_size
        folds.append(ids[start:stop])
        current = stop

    for i in range(k):
        val_ids = folds[i]
        train_ids = np.concatenate([folds[j] for j in range(k) if j != i])
        yield train_ids, val_ids
```

```python
print('K-Fold')
print()

X = np.arange(10)
y = np.array([0] * 5 + [1] * 5)

print(f'X = {X}')
print(f'y = {y}')
print()

for i, ids in enumerate(k_fold_cv(X, y, k=5)):
    train_ids, val_ids = ids

    train_balance = np.unique(y[train_ids], return_counts=True)[1] / len(y[train_ids])
    val_balance = np.unique(y[val_ids], return_counts=True)[1] / len(y[val_ids])

    print(f'Fold {i + 1}:')
    print(f'Split: {train_ids} {val_ids} ')
    print(f'Balance: {train_balance} {val_balance}')
    print()
```

```python
def stratified_k_fold_cv(X, y, k=5):

    n_samples = len(X)
    values, y_ids = np.unique(y, return_inverse=True)
    value_counts = np.bincount(y_ids)
    fold_sizes = np.zeros((len(values), k), dtype=int)

    for i in range(len(values)):
        fold_sizes[i] = (value_counts[i] // k) * np.ones(k, dtype=int)
        fold_sizes[i][:value_counts[i] % k] += 1

    folds = {}
    for i in range(len(values)):
        value_ids = np.where(y == values[i])[0]
        np.random.shuffle(value_ids)
        current = 0

        for fold in range(k):
            start, stop = current, current + fold_sizes[i, fold].item()

            if fold not in folds:
                folds[fold] = np.array([], dtype=int)

            folds[fold] = np.concatenate([folds[fold], value_ids[start:stop]])
            current = stop

    for val_fold in range(k):
        val_ids = folds[val_fold]
        train_ids = np.concatenate([folds[train_fold] for train_fold in range(k) if train_fold != val_fold])
        yield train_ids, val_ids
```

```python
print('Stratified k-Fold')
print()

X = np.arange(10)
y = np.array([0] * 5 + [1] * 5)

print(f'X = {X}')
print(f'y = {y}')
print()

for i, ids in enumerate(stratified_k_fold_cv(X, y, k=5)):
    train_ids, val_ids = ids

    train_balance = np.unique(y[train_ids], return_counts=True)[1] / len(y[train_ids])
    val_balance = np.unique(y[val_ids], return_counts=True)[1] / len(y[val_ids])

    print(f'Fold {i + 1}:')
    print(f'Split: {train_ids} {val_ids} ')
    print(f'Balance: {train_balance} {val_balance}')
    print()
```

```python
def leave_one_out_cv(X, y):

    n_samples = len(X)
    ids = np.arange(n_samples)

    for i in range(n_samples):
        val_ids = [i]
        train_ids = np.delete(ids, i)
        yield train_ids, val_ids
```

```python
print('Leave-one-out')
print()

X = np.arange(10)
y = np.array([0] * 5 + [1] * 5)

print(f'X = {X}')
print(f'y = {y}')
print()

for i, ids in enumerate(leave_one_out_cv(X, y)):
    train_ids, val_ids = ids

    train_balance = np.unique(y[train_ids], return_counts=True)[1] / len(y[train_ids])
    val_balance = np.unique(y[val_ids], return_counts=True)[1] / len(y[val_ids])

    print(f'Fold {i + 1}:')
    print(f'Split: {train_ids} {val_ids} ')
    print(f'Balance: {train_balance} {val_balance}')
    print()
```
